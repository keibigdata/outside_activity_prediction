```python
import requests 
import pandas as pd
import time
```

```python
station_info = pd.read_csv('raw_data/subway_station_csv2.csv', encoding='CP949', index_col =False)
```

```python
name_list = station_info["name"]
```

```python
addr_list = pd.DataFrame(columns=["name", "address"])
```

```python
# test code
name = "서울역"
url = "https://maps.googleapis.com/maps/api/place/textsearch/json?key=AIzaSyALGG7xOXPZ7iO6BiWLm0b890gV7aMGVNQ&l&language=ko&resion=KR"
query = []
query.append('한국 ')
query.append(name)
query.append('역')
query = ''.join(query)

params = {'query': query}
cur_html = requests.get(url, params=params).json()
cur_html["status"] = "not"
if(cur_html["status"]=="OK"):
    addr = cur_html['results'][0]['formatted_address']
    newD = [{'name': name, 'address': addr}]
    newDf = pd.DataFrame(newD)
    addr_list = addr_list.append(newDf, ignore_index=True)
    #addr_list.append({"name":name, "address":addr})
else:
    addr_list = addr_list.append(pd.DataFrame([{"name":name, "address":"err"}]))
```

```python
for name in name_list:
    url = "https://maps.googleapis.com/maps/api/place/textsearch/json?key=AIzaSyALGG7xOXPZ7iO6BiWLm0b890gV7aMGVNQ&l&language=ko"
    
    query = []
    query.append('한국 ')
    query.append(name)
    query.append('역')
    query = ''.join(query) 
    
    params = {'query': query}
    cur_html = requests.get(url, params=params).json()
    
    time.sleep(3)
    
    print(cur_html["status"])
    if(cur_html["status"]=="OK"):
        addr = cur_html['results'][0]['formatted_address']
        newD = [{'name': name, 'address': addr}]
        newDf = pd.DataFrame(newD)
        addr_list = addr_list.append(newDf, ignore_index=True)
    else:
        addr_list = addr_list.append(pd.DataFrame([{"name":name, "address":"err"}]), ignore_index=True)
```

```python
print(addr_list)
```

```python
addr_list.to_csv("addressList2.csv", index=None)
```
